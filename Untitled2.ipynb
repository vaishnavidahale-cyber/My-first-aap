```python
def string_statistics():
    # Take input from the user
    user_input = input("Enter a string: ")

    # Initialize counters
    vowels_count = 0
    consonants_count = 0
    spaces_count = 0
    lowercase_count = 0

    # Define vowels
    vowels = "aeiouAEIOU"

    # Traverse through each character in the string
    for char in user_input:
        # Check if the character is a vowel
        if char in vowels:
            vowels_count += 1
        # Check if the character is a consonant (alphabetic but not a vowel)
        elif char.isalpha():
            consonants_count += 1
        # Check if the character is a space
        elif char == " ":
            spaces_count += 1
        # Check if the character is lowercase
        if char.islower():
            lowercase_count += 1

    # Print the results
    print(f"Number of Vowels: {vowels_count}")
    print(f"Number of Consonants: {consonants_count}")
    print(f"Number of Spaces: {spaces_count}")
    print(f"Number of Lowercase Letters: {lowercase_count}")

# Call the function to run the program
string_statistics()
```

```python
def capitalize_lines():
    # Take multiple lines of input from the user
    lines = input("Enter a sentence: ")

    # Capitalize all characters in the string
    capitalized_lines = lines.upper()

    # Print the capitalized sentence
    print(capitalized_lines)

# Call the function to run the program
capitalize_lines()
```
