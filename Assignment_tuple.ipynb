```python
# Input from user
user_input = input("Enter a series of integers separated by spaces: ")
tuple_data = tuple(map(int, user_input.split()))

# Print the total number of items in the tuple
print(f"Total number of items in the tuple: {len(tuple_data)}")
```

```python
# Print the last item in the tuple
print(f"The last item in the tuple is: {tuple_data[-1]}")
```

```python
# Print the tuple elements in reverse order
print(f"Tuple elements in reverse order: {tuple_data[::-1]}")
```

```python
# Check if the tuple contains the number 5
if 5 in tuple_data:
    print("Yes, the tuple contains an integer 5.")
else:
    print("No, the tuple does not contain an integer 5.")
```

```python
# Remove the first and last item, then sort the remaining items
modified_tuple = tuple_data[1:-1]
sorted_tuple = sorted(modified_tuple)

# Print the sorted result
print(f"Sorted tuple without the first and last elements: {sorted_tuple}")
```

```python
# Input the prices of sold items
prices_input = input("Enter the prices of sold items separated by spaces: ")
prices_tuple = tuple(map(float, prices_input.split()))

# Print the total number of items sold
print(f"Total number of items sold: {len(prices_tuple)}")
```

```python
# Print the price of the cheapest item sold
print(f"The price of the cheapest item sold: {min(prices_tuple)}")
```

```python
# Print the price of the costliest item sold
print(f"The price of the costliest item sold: {max(prices_tuple)}")
```

```python
# Print the price list in ascending order
print(f"Price list in ascending order: {sorted(prices_tuple)}")
```

```python
# Count how many times the costliest item appears in the tuple
costliest_item_count = prices_tuple.count(max(prices_tuple))

# Print the number of costliest items sold
print(f"The number of costliest items sold: {costliest_item_count}")
```
